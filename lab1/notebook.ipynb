```python
import cv2 as cv
import numpy as np
```

```python
img_path = '../imgs/img1.jpg'
brightness_threshold = 150
```

```python
img = cv.imread(img_path)
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img, cmap='gray')

```

```python
brightness_percentage = 100.0 * np.sum(img > brightness_threshold) / (img.shape[0] * img.shape[1])
print(f"pixels percentage, which brightness higher then {brightness_threshold}: {brightness_percentage:.2f} %")
```
