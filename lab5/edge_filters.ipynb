```python

import cv2 as cv
from matplotlib import pyplot as plt
```

```python
image = cv.imread('../imgs/lenna.png', cv.IMREAD_GRAYSCALE)
```

```python
## Sobel

ksizes = [3, 5, 7, 9]
derivatives = [(0, 1), (1, 0), (1, 1)]

ksize_len = len(ksizes)
derivatives_len = len(derivatives)

fig, axes = plt.subplots(ksize_len + 1, derivatives_len + 1, constrained_layout=True)
fig.set_size_inches(8, 11)
fig.suptitle('Sobel')

axes[0][0].imshow(image, cmap='gray')

for i in range(ksize_len + 1):
    axes[i][0].imshow(image, cmap='gray')

for i in range(derivatives_len + 1):
    axes[0][i].imshow(image, cmap='gray')

for i, ksize in enumerate(ksizes, start=1):
    for j, (dx, dy) in enumerate(derivatives, start=1):
        sobel_image = cv.Sobel(image, ddepth=cv.CV_8U, dx=dx, dy=dy, ksize=ksize)
        axes[i][j].set_title(f'dx={dx}, dy={dy}, ksize={ksize}')
        axes[i][j].imshow(sobel_image, cmap='grey')

fig.show()
```

```python
# Laplacian
ksize_len = len(ksizes)
rows = 2
fig, axes = plt.subplots(ksize_len, rows, constrained_layout=True)
fig.set_size_inches(8, 11)
fig.suptitle('Laplacian')

axes[0][0].imshow(image, cmap='gray')

for i in range(ksize_len):
    axes[i][0].imshow(image, cmap='gray')

for i in range(rows):
    axes[0][i].imshow(image, cmap='gray')

for i, ksize in enumerate(ksizes):
    laplacian_img = cv.Laplacian(image, ddepth=cv.CV_8U, ksize=ksize)
    axes[i][1].set_title(f'ksize={ksize}')
    axes[i][1].imshow(laplacian_img, cmap='grey')

fig.show()
```

```python
## Canny
thrs1 = [25, 50, 75]
thrs2 = [100, 150, 200]

thrs1_len = len(thrs1)
thrs2_len = len(thrs2)

fig, axes = plt.subplots(thrs1_len + 1, thrs2_len + 1, constrained_layout=True)
fig.set_size_inches(8, 11)
fig.suptitle('Canny')

axes[0][0].imshow(image, cmap='gray')

for i in range(thrs1_len + 1):
    axes[i][0].imshow(image, cmap='gray')

for i in range(thrs2_len + 1):
    axes[0][i].imshow(image, cmap='gray')

for i, thr1 in enumerate(thrs1, start=1):
    for j, thr2 in enumerate(thrs2, start=1):
        sobel_image = cv.Canny(image, threshold1=thr1, threshold2=thr2)
        axes[i][j].set_title(f'thr1={thr1}, thr2={thr2}')
        axes[i][j].imshow(sobel_image, cmap='grey')

fig.show()
```
