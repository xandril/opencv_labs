```python
import cv2 as cv
import numpy as np
```

```python
img_path = '../imgs/img2.jpg'
img = cv.imread(img_path)
img = cv.cvtColor(img, cv.COLOR_RGB2BGR)

```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.show()
```

```python
cv.circle(img=img,
          center=(40, 60),
          radius=8,
          color=(255, 0, 0),
          thickness=1)  # 
cv.ellipse(img=img, 
           center=(30, 50), 
           color=(255, 255, 0), 
           thickness=1, 
           angle=20, 
           startAngle=0, 
           endAngle=360,
           axes=(5, 10))  #  

cv.ellipse(img=img, 
           center=(60, 40), 
           color=(255, 255, 0), 
           thickness=1, 
           angle=20, 
           startAngle=0,
           endAngle=360,
           axes=(7, 16))  #  

points = np.array([[0, 90], [13, 73], [40, 110]], dtype=np.int32)
cv.polylines(img=img, 
             pts=[points], 
             isClosed=True, 
             color=(255, 0, 0))  #  

points = np.array([[100, 27], [155, 22], [140, 60]], dtype=np.int32)
cv.polylines(img=img, 
             pts=[points], 
             isClosed=True, 
             color=(255, 0, 0))  #  
```
